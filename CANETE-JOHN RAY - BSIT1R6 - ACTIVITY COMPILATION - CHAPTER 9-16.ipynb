```python
print("5. Design a class to represent a playing card and another one to represent a deck of cards. Using these two classes, implement your favorite card game.\n\n")

import random

class Card():
    def __init__(self, suit, val):
        self.suit = suit
        self.val = val

    def show(self):
        print("{} of {}".format(self.val, self.suit))

class Deck():
    def __init__(self):
        self.cards = []
        self.build()

    def build(self):
        for c in ["Spade", "Club", "Diamonds", "Hearts"]:
            for n in range(1, 14):
                self.cards.append(Card(c, n))

    def print(self):
        for x in self.cards:
            x.show()

    def shuffle(self):
        for i in range(len(self.cards) -1, 0, -1):
            r = random.randint(0, i)
            self.cards[i], self.cards[r] = self.cards[r], self.cards[i]

    def draw(self):
        return self.cards.pop()

class player:
    def __init__(self, name):
        self.name = name
        self.hand = []

    def draw(self, deck):
        self.hand.append(deck.draw())
        return self


    def showhands(self):
        for card in self.hand:
            card.show()

deck = Deck()
deck.shuffle()

john = player('john')
john.draw(deck)
john.showhands()

```

```python
print("8. Implement a queue using linked lists.\n")
class Node():
    def __init__(self, data):
        self.data = data
        self.next = None
        
class Queue:
    def __init__(self):
        self.front = None
        self.rear = None
        
    def Enqueue(self, data):
        if self.rear is None:
            self.front = self.rear = Node(data)
        else:
            self.rear.next = Node(data)
            self.rear = self.rear.next
            
    def Dequeue(self):
        if self.front is None:
            return "Queue is Empty"
        else:
            to_return = self.front.data
            self.front = self.front.next
            return to_return
        
    def IsEmpty(self):
        return self.front is None
        
    def Size(self):
        count = 0 
        cur = self.front
        while(cur):
            count+=1
            cur = cur.next
        return count
    
    def Front(self):
        return self.front.data
    
    def Rear(self):
        return self.rear.data
    
q = Queue()
q.Enqueue(1)
q.Enqueue(2)
q.Enqueue(3)
q.Enqueue(4)
```

```python
print("9. Write a recursive function to compute the factorial of a number.\n\n")
def recursive_factorial(n):
   if n == 1:
       return n
   else:
       return n*recursive_factorial(n-1)

number = 5


if number < 0:
   print("Sorry, factorial does not exist for negative numbers")
elif number == 0:
   print("The factorial of 0 is 1")
else:
   print(f"The factorial of", number, "is", recursive_factorial(number))
```

```python
print("10. Set up a random experiment to test the difference between a sequential search and a binary search on a list of integers.\n")

def binary_search(number_of_list, number_to_find):
    left_index = 0
    right_index = len(number_of_list) - 1
    mid_index = 0

    while left_index <= right_index:
        mid_index = (left_index + right_index) // 2
        mid_number = number_of_list[mid_index]

        if mid_number == number_to_find:
            return mid_index

        if mid_number < number_to_find:
            left_index = mid_index + 1
        else:
            right_index = mid_index - 1

    return -1


def sequentialSearch(alist, item):
    pos = 0
    found = False

    while pos < len(alist) and not found:
        if alist[pos] == item:
            found = True
        else:
            pos = pos+1

    return found




if __name__ == "__main__":
    number_of_list=[2, 4, 6, 45, 67, 73, 78]
    number_to_find = 78

    index = binary_search(number_of_list, number_to_find)
    print(f"Number found at index {index} using binary search")
    print(sequentialSearch(number_of_list, 4))
    print(sequentialSearch(number_of_list, 56))
    
print("\nIn binary search it will return the index number while on the sequential search return true if the element found in the list")
print("and false if it is not found. The difference between the two is that from the word Bi in binary means divide into two while")
print("the seqeuntial, it goes to the whole elements sqeuntially until you find the match. So the better one is the binary search")
```

```python
print("11. Suppose you are a computer scientist/art thief who has broken into a major art gallery. All you have with you to haul out your stolen art is your knapsack which only holds W pounds of art, but for every piece of art you know its value and its weight. Write a dynamic programming function to help you maximize your profit. Here is a sample problem for you to use to get started: Suppose your knapsack can hold a total weight of 20. You have 5 items as follows:\n\n")

def KS(total_capacity, weight, value, n):

    if n == 0 or total_capacity == 0:
        return 0

    if (weight[n - 1] > total_capacity):
        return KS(total_capacity, weight, value, n - 1)

    else:
        return max(value[n - 1] + KS(total_capacity - weight[n - 1], weight, value, n - 1), KS(total_capacity, weight, value, n - 1))



value = [3,4,8,8,10]
weight = [2,3,4,5,9]
total_capacity = 20
n = len(value)
print("The answer is: ",KS(total_capacity, weight, value, n))
```
